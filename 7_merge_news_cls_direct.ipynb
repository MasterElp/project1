```python
import pandas as pd
import numpy as np
folder = 'd:/git/project1/generated'
```

```python
#btc_direct_table = pd.read_csv(f'{folder}/btc_daily_change_direct.csv')
btc_direct_table = pd.read_csv(f'{folder}/bts_price_changes.csv')
btc_direct_table
```

```python
date_time_title_short = pd.read_csv(f'{folder}/date_time_title_short.csv')
```

```python
btc_timestamps = btc_direct_table['timestamp']
#btc_direct =  btc_direct_table['direct']
btc_direct =  btc_direct_table['direct_30']
news_timestamps_all = date_time_title_short['date']
```

```python
def get_directs(news_timestamps):
    directs = []
    for news_timestamp in news_timestamps:
        #print(ts_news)
        finded = False
        for i in range(btc_direct.shape[0]):
            if news_timestamp == btc_timestamps[i]:
                directs.append(btc_direct[i])
                finded = True
                break;
        if (not finded):
            print('not finded')
            directs.append(0)    
    return directs
```

```python
def get_union_table(num):
    table = pd.read_csv(f'{folder}/classification_{num*1000}_{(num+1)*1000-1}.csv')
    table['timestamp'] = np.array(news_timestamps_all[num*1000:(num+1)*1000-1])
    table['direct'] = get_directs(table['timestamp'])
    return table
    
```

```python
%%time
res_table = pd.DataFrame(None)
res_table = res_table.append(get_union_table(0), sort=False, ignore_index=True)
res_table
```

```python
%%time
start = 0
for j in range(2):#max=60
    res_table = pd.DataFrame(None)#clear_table
    for i in range(10):
        num = i+j*10+start
        print(num)
        res_table = res_table.append(get_union_table(num), sort=False, ignore_index=True)
    res_table.to_csv(f'{folder}\cls_direct30_{num}.csv', index=False)
```
