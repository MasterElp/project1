```python
import pandas as pd
```

```python
folder = 'd:/git/project1/generated'
```

```python
prices = pd.read_csv(f'{folder}/bts_price_changes.csv')
```

```python
res = pd.DataFrame()
```

```python
res['direct'] = prices.apply(lambda x: 1 if (x.Close - x.Open) > 0 else 0, axis=1)
```
