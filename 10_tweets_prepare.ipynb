```python
import time, datetime
df['date'] = df['timestamp'].apply(lambda x:int(time.mktime(datetime.datetime.strptime(x.split()[0], "%Y-%m-%d").timetuple())))
```

```python
def time_to_int(time_str):
    time_str = time_str.split('+')[0]
    hh, mm , ss = map(int, time_str.split(':'))
    return ss + 60*(mm + 60*hh)
df['time'] = df['timestamp'].apply(lambda x:time_to_int(x.split()[1]))
```
