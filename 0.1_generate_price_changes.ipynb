```python
import pandas as pd
import numpy as np
folder = 'd:/git/project1/generated'
```

```python
df = pd.read_csv(f'{folder}/bts_price_changes.csv')
df
```

```python
del df['Unnamed: 0']
df
```

```python
period = 180
df[f'change_percent_{period}'] = df.apply(lambda x: x[f'change_{period}_day']/x['Close'], axis=1)
target = f'change_percent_{period}'
df[['timestamp', 'Open', 'High', 'Low', 'Close', 'Volume', target]].corr()[target].sort_values()
```

```python
df.to_csv(f'{folder}/bts_price_changes.csv', index=False)
```
