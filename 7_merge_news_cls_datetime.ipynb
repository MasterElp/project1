```python
date_time_title_short = pd.read_csv(f'{folder}/date_time_title_short.csv')
```

```python
def get_table(num):
    table = pd.read_csv(f'{folder}/classification_{num*1000}_{(num+1)*1000-1}.csv')
    table['timestamp'] = np.array(date_time_title_short.loc[num*1000:(num+1)*1000-2, 'date'])
    directs = []
    for ts_news in table.timestamp:
        #print(ts_news)
        finded = False
        for i in range(btc_direct.shape[0]):
            #print(btc_direct.loc[i,:])
            if ts_news == btc_direct.loc[i,:].timestamp:
                directs.append(btc_direct.loc[i,:].direct)
                finded = True
                break;
        if (not finded):
            print('not finded')
            directs.append(0)
    table['direct'] = directs
    return table
```

```python
start = 394
for j in range(60):
    res_table = pd.DataFrame(None)
    for i in range(10):
        num = i+j*10+start
        print(num)
        res_table = res_table.append(get_table(num), sort=False, ignore_index=True)
    res_table.to_csv(f'{folder}\set_{num}.csv')
```
